```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
df = pd.read_excel('Junior Data Analyst Task.xlsx', sheet_name='RAW_Task2')
```

```python
df.info()
```

# Preprocessing

```python
#Предположу, что param_1, param_2, param_3, param_4 можно считать как True, False
#поэтому заполняют "пропущенные" значения нулями.

df['param1'].fillna(value=0, inplace=True)
df['param1'] = df['param1'].astype('int64')

df['param2'].fillna(value=0, inplace=True)
df['param2'] = df['param2'].astype('int64')

df['param3'].fillna(value=0, inplace=True)
df['param3'] = df['param3'].astype('int64')

df['param4'].fillna(value=0, inplace=True)
df['param4'] = df['param4'].astype('int64')
```

```python
#Создаём новую переменную: наличие соцсети, проверим далее будет ли такой признак важен
df['has_social'] = df['social'].isna()
```

```python
df['social'].fillna(value='NoSocial', inplace=True)
```

```python
#Разбиваем дату на число и месяц
df['regDateTime'] = pd.to_datetime(df['regDateTime'])
df['regMonth'] = pd.DatetimeIndex(df['regDateTime']).month
df['regDay'] = pd.DatetimeIndex(df['regDateTime']).day
```

```python
df.info()
```

```python
df.columns
```

# Model 1

```python
data1 = df.drop(columns=['email', 'os', 'social', 'regDateTime', 'Revenue'])
```

```python
data1.head()
```

```python
dummy_os = pd.get_dummies(df['os'])
dummy_social = pd.get_dummies(df['social'])

new_df = data1.merge(dummy_os, left_index=True, right_index=True).merge(dummy_social, left_index=True, right_index=True)
new_df.head()
```

```python
df_data = new_df.drop(['Conversion'], axis=1)
df_data.head()
```

```python
labels = new_df['Conversion']
labels.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_data, labels, test_size=0.25, random_state=42)
scaler = MinMaxScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Model 1 - k-nearest neighbors model
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)

#Training Score
print('Training Score:', knn.score(X_train, y_train))

y_pred = knn.predict(X_test)

#Test Score Metrics
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("\n Confusion matrix: \n", metrics.confusion_matrix(y_pred, y_test))
print("\n Classification report: \n", metrics.classification_report(y_pred, y_test))
print("\n ROC Curve:", metrics.roc_auc_score(y_pred, y_test))
```

## Heatmap For Model 1

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data1.corr(), annot=True, fmt='.2f', linewidth=0.5, ax=ax)
plt.show()

#heatmap говорит о том, что некоторые переменные являются не самыми важными для модели, стоит их убрать и попробовать заново
```

# Model 3

```python
data3 = df.drop(columns=['email', 'os', 'social', 'regDateTime', 'Revenue'])
data3.head()
```

```python
dummy_os = pd.get_dummies(df['os'])
new_df3 = data3.merge(dummy_os, left_index=True, right_index=True)
```

```python
df3_data = new_df3.drop(['Conversion'], axis=1)
df3_data.head()
```

```python
labels = new_df3['Conversion']
labels.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df3_data, labels, test_size=0.2, random_state=42)
scaler = MinMaxScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Model 3 - XGBoosts classifier
xgb = XGBClassifier(scale_pos_weight=1,
                      learning_rate=0.01,  
                      colsample_bytree=0.4,
                      subsample=0.8,
                      objective='binary:logistic', 
                      n_estimators=1000, 
                      reg_alpha=0.1,
                      max_depth=3, 
                      gamma=10,
                      random_state=42)

xgb.fit(X_train, y_train)

#Training Score
print('Training Score:', xgb.score(X_train, y_train))

y_pred = xgb.predict(X_test)

#Test Score Metrics
print("\nAccuracy:", metrics.accuracy_score(y_test, y_pred))
print("\nConfusion matrix: \n", metrics.confusion_matrix(y_pred, y_test))
print("\nClassification report: \n", metrics.classification_report(y_pred, y_test))
print("\nROC Curve:", metrics.roc_auc_score(y_pred, y_test))
```

# Feature Importance For Model 3

```python
importance = xgb.feature_importances_

for i,v in enumerate(importance):
    print('Feature: %s, Score: %.5f' % (new_df3.columns[i],v))

plt.bar([x for x in range(len(importance))], importance)
plt.show()

# столбчатая диаграмма говорит о том, что наличие определённой операционной системы является важным признаком для нашей модели, 
# а также что некоторые переменные являются не самыми важными для модели, стоит их убрать и попробовать заново
```
